```python
import numpy as np
import matplotlib.pyplot as plt
```

**пункт А:**

```python
n = 100000
for i in range(5):
    x, y = np.random.randn(n), np.random.randn(n)
    y[np.abs(y) < 1e-12] = 1e-12
    plt.plot(np.cumsum(x / y) / np.arange(1, n + 1))
plt.show()
```

**пункт Б:**

```python
for i in range(5):
    x, y, z = np.random.randn(n), np.random.randn(n), np.random.randn(n)
    g = np.sqrt((y**2 + z**2) / 2)
    g[g < 1e-12] = 1e-12
    plt.plot(np.cumsum(x / g) / np.arange(1, n + 1))
plt.show()
```
