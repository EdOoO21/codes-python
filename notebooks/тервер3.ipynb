```python
import matplotlib.pyplot as plt
from random import randint
```

**пунктик А:** 
Просто будем собирать рандомные числа 10000 раз в массив, пока не выпадет 8 и добавлять длину массива, также сразу посчитаем в скольки случаях длина массива <=4 для пункта В. Ну а потом просто построим график :

```python
y = [0 for _ in range(10000)]
amount = 0
for i in range(10000):
    x = 0
    counter = 0
    while (x != 8):
        x = randint(3,8)
        counter += 1
    if counter <= 4:
        amount += 1
    y[i] = counter

plt.hist(y, bins=100, range=(min(y),max(y)))
plt.xlabel("кол-во бросков до 8")
plt.ylabel("сколько раз понадобилось столько бросков из 10к")
plt.grid(True)
plt.show()
```

**пунктик Б :**
оценим безусловное мат ожидание - в данном случае это сумма всех делить количество :

```python
print(f"оценка безусловного мат ожидания = {sum(y) / 10000}")
```

**пунктик В :** так как из пункта А мы насчитали кол-во случаев, когда кол-во бросков <5, то просто оценим вероятность, поделив это число на кол-во эксперементов :

```python
print(f"оценка вероятности окончания игры до 5 бросков не включительно = {amount / 10000}")
```

**пунктик Г** ну построим как сказали : 

```python
y = [0 for _ in range(10000)]
i = 0
amount = 0
while (i < 10000):
    x = 0
    counter = 0
    while (x != 8):
        x = randint(1,8)
        counter += 1
        if 1 <= x <= 2:
            break
    else:
        if counter <= 4:
            amount += 1
        y[i] = counter
        i += 1

plt.hist(y, bins=50, range=(min(y),max(y)))
plt.xlabel("кол-во бросков до 8")
plt.ylabel("сколько раз понадобилось столько бросков из 10к")
plt.grid(True)
plt.show()
```

**пунктик Д** посчитаем как в Б :

```python
print(f"условное мат ожидание с учетом условия = {sum(y) / 10000}")
```
