```python
def run_simulation(days, not_weekend, weekend):
    ans = np.zeros(10000)
    for i in range(10000):
        sum = 0
        for today in range(days):
            if ((today % 7) == 5) or ((today % 7) == 6):
                now = weekend
            else:
                now = not_weekend
            daily_payouts = np.random.uniform(0, 10**36, np.random.poisson(now))
            sum += daily_payouts.sum()
        ans[i] = sum
    return ans / 10**36
```

**Пунктик А:**

```python
plt.figure(figsize=(20, 15))
plt.hist(run_simulation(10, 100 / 30, 100 / 30), bins=100)
plt.xlabel("выплата")
plt.ylabel("частота")
plt.grid()
plt.show()
```

**Пунктик Б:**

```python
res = run_simulation(10, 100 / 30, 100 / 30)
print(sum([x > 12 for x in res]) / len(res))
```

**Пунктик В:**

```python
res = run_simulation(30, 100 / 30, 100 / 30)
sorts = sorted(res)
ans = int(0.95 * len(sorts)) - 1
print(sorts[ans])
```
