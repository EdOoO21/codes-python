В самом начале утвердим, что если число от 1 до 30 включительно, то это равносильно тому, что задача была решена в ретроградный меркурий

```python
import random
from math import comb
```

**A:** Сделаем 10к выборок и сразу найдем, сколько раз Snew <= S :

```python
Snew = 0
S = 5
amount_of_true = 0
arr = []
for i in range(10000):
    Snew = 0
    arr = random.sample(range(1,51), 11)
    for x in arr:
        if x <= 30:
            Snew += 1
    print(f"выборка номер {i + 1} Snew = {Snew} || верные =", *arr)
    if Snew <= S:
        amount_of_true += 1
```

**B:** Оценим p_value :

```python
print(f"p_value = {amount_of_true / 10000}")
```

**С:** Так как перебрать и вывести все выборки очень долго (их много), то просто посчитаем втупую сколько всего перестановок и сколько подходит под условие Snew <= S, а потом поделим :

```python
# кол-во всех выборок
total_amount = comb(50, 11)
answer = 0

# найдем удачные выборки, когда 5 и менее значения <= 30
for i in range(6):
    answer += comb(30, i) * comb(20, 11 - i)

print(f"точное p_value = {answer / total_amount}")
```

**D:** отвергаем или нет :

```python
if answer < 0.05 :
  print("отвергаем гипотезу H0")
else:
  print("не отвергаем гипотезу H0")
```
