a) Рисуем пять траекторий Rn

```python
import numpy as np
import matplotlib.pyplot as plt

N = 100000

plt.figure(figsize=(10, 6))
for i in range(5):
    X = np.random.randn(N)
    Y = np.random.randn(N)

    Y[np.abs(Y) < 1e-12] = 1e-12

    R = X / Y
    Rt = np.cumsum(R) / np.arange(1, N + 1)

    plt.plot(Rt, alpha=0.7)

plt.title(r'Накопленные средние $\overline{R}_n$ для 5 траекторий')
plt.xlabel('n')
plt.ylabel(r'$\overline{R}_n$')
plt.show()
```

б) Пять траекторий Ln

```python
plt.figure(figsize=(10, 6))
for i in range(5):

    X = np.random.randn(N)
    Y = np.random.randn(N)
    Z = np.random.randn(N)

    d = np.sqrt((Y**2 + Z**2) / 2)
    d[d < 1e-12] = 1e-12

    L = X / d
    Lt = np.cumsum(L) / np.arange(1, N + 1)

    plt.plot(Lt, alpha=0.7)

plt.title(r'Накопленные средние $\overline{L}_n$ для 5 траекторий')
plt.xlabel('n')
plt.ylabel(r'$\overline{L}_n$')
plt.show()
```
