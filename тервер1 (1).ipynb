```python
import csv
import random
from math import factorial
file = "https://raw.githubusercontent.com/bdemeshev/hse_pmi_probability_2024_2025/main/home_assignments/ha01_data.csv"
import pandas as pd
table = pd.read_csv(file)
```

табличка :

```python
print(table)
```

пунктик А :

```python
a = sum([table["y"][i] for i in range(0,30)]) / 30
b = sum([table["y"][i] for i in range(30,40)]) / 10
s = b - a
print(f" S = {s}")
```

пунктик Б и сразу посчитаем в скольки случаях у нас S_new <= S:

```python
all_sum = (a * 30) + (b * 10)
counter = 0
arr = []
for i in range(1,10001):
  arr = random.sample(range(0,40), 10)
  b = sum([table["y"][x] for x in arr])
  a = (all_sum - b) / 30
  b /= 10
  if b - a <= s :
    counter += 1
  print(f"перестановка {i} s = {b - a}")
print(f"кол-во случаев, когда S_new <= S : {counter}")
```

пунктик B :

```python
p_value = counter / 10000
print(f"Если S_new <= S в {counter} случаях, то p-value = {p_value}")
```

пунктик Д :

```python
if p_value < 0.05 :
  print("достаточное кол-во людей переставали смотреть рилсики перед дд, респект работягам, отвергаем гипотезу H0")
else:
  print("в слишком малом кол-ве случаев кол-во просмотренных рилсов изменилось при приблежении дедлайна, всем пофиг походу вообще, не отвергаем гипотезу H0")
```
